# K means 
K-means is an unsupervised machine learning algorithm technique that allows us to cluster data points, and allows us to find patterns in the data that would help us analyze it more effectively. It's an itterative algorithm meaning it will converge to the optimal clustering over time.

### To run a K-means clustering :
- specify the number of clusters you want ( usually referred as k)
- randomly initialize the centroid for each cluster. The centroid is the data-point at the center of the cluster
- then it determines which data points belong to which cluster by finding the closest centroid to each data point
- then it updates the centroids based on the geometric mean of all the data points of the cluster
- run 3,4 untill the centroids stop changing. Each run here is an iteration.

```python
import pandas as pd
import numpy as np
```

```python
players = pd.read_csv('players_22.csv')
```

```python
players
```

```python
# features would be using for clustering
features=["overall","potential","wage_eur","value_eur","age"]
```

```python
players=players.dropna(subset=features) # drop any row if any of the following columns has any missing values
```

```python
data = players[features].copy()
```

```python
data
```

### Rescaling

```python
data = (data-data.min())/(data.max()-data.min()) *9+1 
```

We did min-max scaling here and scalled it from 0-9 then we added 1 to make it range from 1-10

```python
data.describe() # min would now be 1 and max be 10
```

Hence we made sure that no one column will dominate the other and also preserved the variance of the column values

```python
data.head(3)
```

### Initialize the random centroids

```python
# Initializing our random centroids
# it would be similar to a row here and it would have values for each column, though would be random
# centroid = data.apply(lambda x: float(x.sample()))
```

The apply method would iterrate through all the columns in the dataframe and then x.sample() -- select a random value from each column and then turn it into float.

```python
#centroid
```

The above is one centroid, which has random initialised values for each variable we want to cluster according to. Now we basically want to have k-such cluster centroids so we loop.

```python
type(centroid)
```

```python
def random_centroids (data,k):
    centroids=[]
    for i in range(k):
        centroid = data.apply(lambda x: float(x.sample()))
        centroids.append(centroid)
    return pd.concat(centroids, axis=1) # as each centroid is a pandas series and concat converts it into a pandass dataframe
```

```python
centroids = random_centroids(data,5)
```

```python
centroids
```

### Labeling each data-point 

```python
# euclidean distance between data-point and cluster center, and then find the cluter center which is the closest to it
# and then assign the data point to that cluster
# np.sqrt(((data-centroids.iloc[:,0])**2).sum(axis=1))
# we will then apply this to all of our centroids by appply to all columns using lambda function
distances= centroids.apply(lambda x :np.sqrt(((data- x)**2).sum(axis=1))) # here x is a column
```

```python
distances
```

```python
# then we need to find which cluster people are in, we can do it by .idxmin(axis=1)
# index of the minimum value
distances.idxmin(axis=1)
```

```python
# function
def get_labels(data, centroids):
    distances= centroids.apply(lambda x :np.sqrt(((data- x)**2).sum(axis=1)))
    return distances.idxmin(axis=1)
```

```python
labels = get_labels(data, centroids)
```

```python
labels
```

```python
# to see how many players are in each cluster
labels.value_counts()
```

### Updating of clusters

###### calculation of geometric mean

```python
#np.exp(np.log(x).mean()) # this would be the GM as multiplying each and storing would consume memory
```

1. We are taking log of all the values and then and then the mean of it (for each row)
2. then we are raising it to exponent to get the GM
3. We would apply this t0 each of the rows

```python
data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
# it splits our data by labels and then we apply the function to each label group
```

The above is the GM of all of the values for players assigned to each clusters, then we need to take transpose of it as initially we had columns as clusters and rows as labels, similar to centroids. These are the new centroids.

```python
def new_centroids(data, labels, k):
    return data.groupby(labels).apply(lambda x: np.exp(np.log(x).mean())).T
```

### Looping and updating the cluster assignment and also visualizing it

```python
from sklearn.decomposition import PCA # PCA stands for Principal Component Analysis
import matplotlib.pyplot as plt # plotting
from IPython.display import clear_output # clear the output each time we plotted, so that we can plot new one
```

PCA would help us visualise our data better as we have k-dimensional data

```python
# plotting the clusters
def plot_clusters(data, labels, centroids, iteration):
    pca = PCA(n_components=2) # only 2d we want
    data_2d=pca.fit_transform(data) # firsts fits in to our data and then transforms our data
    centroids_2d = pca.transform(centroids.T) # the pca expects our rows to be data points and columns to be labels
    
    #ploting
    clear_output(wait=True) # would clear the plot we have so that we can plot a new one on top of it
    plt.title(f'Iteration{iteration}')
    plt.scatter(x=data_2d[:,0], y=data_2d[:,1], c=labels) # 0,1 are the ones returned by pca
    plt.scatter(x=centroids_2d[:,0], y=centroids_2d[:,1])
    plt.show()
```

The above is to visualise the cluster shifting as the iteration proceeds

```python
max_iterations=200
k=4

centroids= random_centroids(data,k)
old_centroids = pd.DataFrame()
iteration =1

while iteration < max_iterations and not centroids.equals(old_centroids): 
# it tells to stop the loop if iteraton exceeds the max iterations ot centroids=old centroids
    old_centroids = centroids
    labels = get_labels(data, centroids)
    centroids = new_centroids(data, labels, k)
    plot_clusters(data, labels, centroids, iteration)
    iteration+=1
```
